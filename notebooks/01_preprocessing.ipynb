# Data Preprocessing

## Objective:
- Load raw dataset
- Explore data
- Handle missing values & duplicates
- Normalize numerical features
- Save cleaned dataset for further analysis

```python
# Import required libraries
import sys
import pandas as pd

# Add src folder to path so we can import our modules
sys.path.append('../src')

# Import preprocessing functions
from preprocessing import load_data, explore_data, clean_data, normalize_data, save_data, encode_gender
```

```python
# Define file path
data_path = '../data/raw/mall_customers.csv'

# Load dataset
df = load_data(data_path)

# Display first 5 rows
df.head(20)
```

```python
df= encode_gender(df)
df.head()
```

```python
# Understand dataset structure
explore_data(df)
```

```python
# Remove missing values and duplicates
df = clean_data(df)

# Check again after cleaning
print("After Cleaning:")
print(df.isnull().sum())
```

```python
# Normalize Income and Spending Score
df = normalize_data(df)

# View updated dataset
df.head()
```

```python
# Define output path
output_path = '../data/processed/clean_data.csv'

# Save cleaned dataset
save_data(df, output_path)

print("✅ Clean data saved successfully!")
```
