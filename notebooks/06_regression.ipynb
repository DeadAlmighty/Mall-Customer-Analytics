# Regression Analysis

## Objective:
- Predict customer spending score based on annual income
- Understand relationship between variables

## Why Regression?
- Used for predicting continuous values
- Helps in understanding trends between variables

## Algorithms Used:
- Linear Regression

## Output:
- Regression line
- Prediction model

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load processed dataset
df = pd.read_csv('../data/processed/clean_data.csv')

# -------------------------------
# STEP 1: DEFINE FEATURES & TARGET
# -------------------------------
# Independent variable (input)
X = df[['Annual Income (k$)']]

# Dependent variable (output)
y = df['Spending Score (1-100)']

# -------------------------------
# STEP 2: TRAIN MODEL
# -------------------------------
model = LinearRegression()
model.fit(X, y)

# -------------------------------
# STEP 3: MAKE PREDICTIONS
# -------------------------------
y_pred = model.predict(X)

# -------------------------------
# STEP 4: EVALUATE MODEL
# -------------------------------
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("Mean Squared Error:", mse)
print("R2 Score:", r2)

# -------------------------------
# STEP 5: VISUALIZE REGRESSION
# -------------------------------
plt.figure(figsize=(6,4))

# Scatter plot of actual data
plt.scatter(X, y, color='blue', label='Actual Data')

# Regression line
plt.plot(X, y_pred, color='red', label='Regression Line')

plt.title("Income vs Spending Score (Regression)")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()

# Save plot
plt.savefig('../outputs/plots/regression_plot.png')

plt.show()

# -------------------------------
# STEP 6: SAVE MODEL (OPTIONAL)
# -------------------------------
import joblib
joblib.dump(model, '../outputs/models/regression_model.pkl')

print("\nRegression Analysis Completed Successfully ✅")
```
