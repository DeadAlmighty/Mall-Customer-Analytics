# Principal Component Analysis (PCA)

## Objective:
- Reduce dimensionality of the dataset
- Preserve maximum variance in fewer components
- Improve efficiency of data mining algorithms

## Why PCA?
- Removes redundancy in data
- Helps in visualization
- Improves performance of models

## Applied On:
- Annual Income
- Spending Score

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load processed dataset
df = pd.read_csv('../data/processed/clean_data.csv')

# Select numerical features for PCA
features = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Standardize data (IMPORTANT for PCA)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(features)

# Apply PCA (reduce to 2 components)
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Create new dataframe with PCA results
pca_df = pd.DataFrame(
    data=pca_result,
    columns=['PC1', 'PC2']
)

# Display explained variance
print("Explained Variance Ratio:", pca.explained_variance_ratio_)

# Plot PCA results
plt.figure(figsize=(6,4))
plt.scatter(pca_df['PC1'], pca_df['PC2'])
plt.title("PCA Result (2 Components)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

# Save plot
plt.savefig('../outputs/plots/pca_plot.png')

plt.show()

# Save PCA dataset (optional but useful)
pca_df.to_csv('../data/processed/pca_data.csv', index=False)

print("\nPCA Completed Successfully ✅")
```
