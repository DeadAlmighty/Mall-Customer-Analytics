# Data Warehouse and OLAP Analysis

## Objective:
- Design a data warehouse model (Star Schema)
- Perform OLAP operations on the dataset
- Analyze data from multiple perspectives

## Concepts Covered:
- Data Warehouse
- Star Schema
- Fact Table & Dimension Tables
- OLAP Operations (Roll-up, Drill-down, Slice, Dice)

## Why Important?
- Core part of Data Warehousing syllabus
- Helps in decision-making and data summarization

```python
# Import required libraries
import pandas as pd

# Load clustered dataset (final processed data)
df = pd.read_csv('../data/processed/clustered_data.csv')

# -------------------------------
# STEP 1: CREATE FACT TABLE
# -------------------------------
# Fact table contains measurable values

fact_table = df[['CustomerID', 'Annual Income (k$)', 'Spending Score (1-100)', 'Cluster']]

print("Fact Table Preview:")
print(fact_table.head())

# -------------------------------
# STEP 2: CREATE DIMENSION TABLES
# -------------------------------

# Dimension 1: Age
dim_age = df[['CustomerID', 'Age']]

# Dimension 2: Gender
dim_gender = df[['CustomerID', 'Gender']]

# Dimension 3: Income Category
df['Income_Category'] = pd.cut(
    df['Annual Income (k$)'],
    bins=3,
    labels=['Low', 'Medium', 'High']
)

dim_income = df[['CustomerID', 'Income_Category']]

print("\nDimension Tables Created")

# -------------------------------
# STEP 3: OLAP OPERATIONS
# -------------------------------

# 🔹 ROLL-UP: Average spending by Gender
rollup = df.groupby('Gender')['Spending Score (1-100)'].mean()
print("\nROLL-UP (Avg Spending by Gender):")
print(rollup)

# 🔹 DRILL-DOWN: Detailed view by Gender and Cluster
drilldown = df.groupby(['Gender', 'Cluster'])['Spending Score (1-100)'].mean()
print("\nDRILL-DOWN (Gender + Cluster):")
print(drilldown)

# 🔹 SLICE: Filter only Female customers
slice_data = df[df['Gender'] == 'Female']
print("\nSLICE (Female Customers):")
print(slice_data.head())

# 🔹 DICE: Filter High Income & High Spending
dice_data = df[
    (df['Income_Category'] == 'High') &
    (df['Spending Score (1-100)'] > 0.7)
]
print("\nDICE (High Income & High Spending):")
print(dice_data.head())

# -------------------------------
# STEP 4: SAVE TABLES
# -------------------------------
fact_table.to_csv('../outputs/reports/fact_table.csv', index=False)
dim_age.to_csv('../outputs/reports/dim_age.csv', index=False)
dim_gender.to_csv('../outputs/reports/dim_gender.csv', index=False)
dim_income.to_csv('../outputs/reports/dim_income.csv', index=False)

print("\nData Warehouse and OLAP Completed Successfully ✅")
```
